# 영화 추천 시스템 - 아이템 기반(Item-based) 최근접 이웃(Nearest Neighbor) 협업 필터링(Collaborative Filtering)
- `MovieLens Latest Datasets`
    - https://grouplens.org/datasets/movielens/latest/
- 영화의 평점을 매긴 사용자와 평점 행렬 등의 데이터이며, 이 중 1MB짜리 small 데이터 사용
- '이 상품을 선택한 다른 고객들은 다음 상품도 구매했음'을 기반으로 추천
    - 사용자들이 어떤 아이템을 좋아하는지/싫어하는지 평가한 척도가 유사한 아이템을 추천
    - 즉, 각 아이템에 대해 사람들이 비슷하게 평가한 정도가 유사도가 되어 그에 기반한 추천

## 데이터 읽기

```python
# 데이터 읽기
import pandas as pd
import numpy as np

movies = pd.read_csv('./MovieLens Latest Datasets/movies.csv')
ratings = pd.read_csv('./MovieLens Latest Datasets/ratings.csv')
print(movies.shape)
print(ratings.shape)
```

```python
movies.head() # 데이터셋 movies에는 영화 제목과 장르
```

```python
ratings.head() # ratings에는 영화 평점이 사용자별로 영화별로 존재한다
```

## 데이터 전처리

```python
# raw 데이터 정리
ratings = ratings[['userId','movieId','rating']] # userid, movieid, rating만
ratings_matrix = ratings.pivot_table('rating', index='userId', columns='movieId')
ratings_matrix.head()
```

```python
# ratings와 movie를 movieID 기준으로 결합
rating_movies = pd.merge(ratings, movies, on='movieId')
rating_movies.head()
```

```python
# 그럼 다시 pivot_table을 이용해서 정리
ratings_matrix = rating_movies.pivot_table('rating', index='userId', columns='title')
ratings_matrix.head()
```

- NaN 값이 많은 희소행렬 형태이다
- 최소 평점이 0.5이므로 NaN값은 모두 0으로 변환하기로 한다

```python
# nan은 0으로 변환
ratings_matrix = ratings_matrix.fillna(0)
ratings_matrix.head()
```

- 이제 사용자-영화 평점 행렬 데이터 세트를 이용해 영화간의 유사도를 측정한다
- 그런데 위 데이터프레임은 userid가 기준인 행 레벨 데이터이므로 유사도를 바로 측정하면 사용자 간의 유사도가 만들어진다. 그러므로 행과 열 위치를 변경해 계산해야한다.

```python
# 유사도 측정을 위해 행렬의 transpose
ratings_matrix_T = ratings_matrix.transpose()
ratings_matrix_T.head()
```

## 코사인 유사도 측정

```python
# 유사도 측정
from sklearn.metrics.pairwise import cosine_similarity

item_sim = cosine_similarity(ratings_matrix_T, ratings_matrix_T)
item_sim_df = pd.DataFrame(data=item_sim, index=ratings_matrix.columns, columns=ratings_matrix.columns)
print(item_sim_df.shape)
item_sim_df.head()
```

## 추천하는/유사한 영화 찾기

```python
# '대부'와 유사한 영화는?
item_sim_df['Godfather, The (1972)'].sort_values(ascending=False)[:5]
```

```python
# '인셉션'과 유사한 영화는?
item_sim_df['Inception (2010)'].sort_values(ascending=False)[1:5]
```
